```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
from bs4 import BeautifulSoup
import requests
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
resp = requests.get("https://stockmarketmba.com/stocksinthesp500.php")
soup = BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'id': 'ipos'})
rows = table.find('tbody').find_all('tr')

tickers = []
industries = []

for row in rows:
    ticker = row.find_all('td')[0].text.strip()
    tickers.append(ticker)
    industry = row.find_all('td')[4].text.strip()
    industries.append(industry)

df_list = []
no_data = []

for i in range(len(tickers)):
    try:
        temp = pdr.get_data_yahoo(tickers[i], start='2011-01-01', end='2020-12-31')
        temp['symbol'] = tickers[i]
        temp['industry'] = industries[i]
        df_list.append(temp)
    except:
        no_data.append(tickers[i])

main_df = pd.concat(df_list)
main_df.to_csv("SP500 stock OHLC.csv")
print(main_df)
```

```python
temp = pdr.get_data_yahoo('BRK-B', start='2011-01-01', end='2020-12-31')
temp['symbol'] = 'BRK-B'
temp['industry'] = industries[tickers.index('BRK.B')]
df_list.append(temp)
main_df = pd.concat(df_list)
main_df['industry'] = np.where(main_df.industry=='Information technology', 'Information Technology', main_df.industry)
main_df.to_csv("SP500 stock OHLC.csv")
print(main_df)
```

```python
num = 10
for i in range(num):
    temp = main_df.iloc[int(len(main_df)*i/10):int(len(main_df)*(i+1)/10), ]
    temp.to_csv("SP500 stock OHLC {}.csv".format(i+1))
```

```python
length = 0
for i in range(num):
    length += len(pd.read_csv("SP500 stock OHLC {}.csv".format(i+1), header=0, index_col=0, parse_dates=[0]))
print(length)
```
