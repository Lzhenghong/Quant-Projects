```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
from bs4 import BeautifulSoup
import requests
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
resp = requests.get("https://stockmarketmba.com/stocksinthesp500.php")
soup = BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'id': 'ipos'})
rows = table.find('tbody').find_all('tr')

tickers = []
industries = []

for row in rows:
    ticker = row.find_all('td')[0].text.strip()
    tickers.append(ticker)
    industry = row.find_all('td')[4].text.strip()
    industries.append(industry)

df_list = []
no_data = []

for i in range(len(tickers)):
    try:
        temp = pdr.get_data_yahoo(tickers[i], start='2003-06-01', end='2022-10-31')
        temp['symbol'] = tickers[i]
        temp['industry'] = industries[i]
        df_list.append(temp)
    except:
        no_data.append(tickers[i])

main_df = pd.concat(df_list)
main_df.to_csv("SP500 stock OHLC extended.csv")
print(main_df)
```

```python
main_df['industry'] = np.where(main_df.industry=='Information technology', 'Information Technology', main_df.industry)
main_df.industry.unique()
```

```python
initial_ticker = main_df.loc[main_df.index == main_df.index.unique()[0]].symbol.unique()
last_month_df = main_df.loc[main_df.index == main_df.index.unique()[-1]]
survived_tickers = last_month_df.loc[last_month_df.symbol.isin(initial_ticker)].symbol.unique()
len(survived_tickers)
```

```python
main_df = main_df.loc[main_df.symbol.isin(survived_tickers)]
main_df
```
